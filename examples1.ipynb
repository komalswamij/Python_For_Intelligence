```python
# Reverse Number for Negative Values
n = int(input("Enter number: "))

sign = -1 if n < 0 else 1
n = abs(n)

reverse = 0

while n > 0:
    digit = n % 10
    reverse = reverse * 10 + digit
    n = n // 10

print("Reverse =", reverse * sign)
```

```python
#Prime Number Check
n = int(input("Enter number: "))
if n > 1:
    for i in range(2, int(n**0.5) + 1):

        if n % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
else:
    print("Not Prime")
```

```python
# Print Prime Numbers from 1 to n
n = int(input("Enter value of n: "))
for num in range(2, n + 1):
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num)
```

```python
# Armstrong Number: Digits ki power ka total = original number
#153 = 1³+5³+3³ = 1+ 125+ 27
n = int(input("Enter number: "))
temp = n
power = len(str(n))
total = 0
while temp > 0:
    digit = temp % 10
    total += digit ** power
    temp //= 10
if total == n:
    print("Armstrong Number")
else:
    print("Not Armstrong Number")
```
